```python
mycursor = mydb.cursor()
mycursor.execute("DROP TABLE IF EXISTS emp_dep; ")
mycursor.execute("DROP TABLE IF EXISTS emp_non_dep ")
```

```python
EMP_Dep_sql="""\
Create table EMP_Dep as
Select ENO, Dept
From EMP;
"""
```

```python
mycursor.execute(EMP_Dep_sql)
```

```python
EMP_Non_Dep_sql="""\
Create table EMP_Non_Dep as
Select ENO, ENAME,TITLE,DOB
From EMP;
"""
```

```python
mycursor.execute(EMP_Non_Dep_sql)
```

```python
EMP_Dep_sql="""\
Select ENO, Dept From EMP_Dep;
"""
EMP_Dep = pd.read_sql(EMP_Dep_sql, con=mydb)
display(EMP_Dep)
```

```python
EMP_Non_Dep_sql="""\
Select ENO, ENAME,TITLE,DOB From EMP_Non_Dep;
"""
EMP_Non_Dep = pd.read_sql(EMP_Non_Dep_sql, con=mydb)
display(EMP_Non_Dep)
```

```python
Constructed_EMP_sql="""\
Select EMP_Dep.ENO,ENAME,TITLE,DOB,EMP_Dep.Dept
From EMP_Dep, EMP_Non_Dep
Where EMP_Dep.ENO = EMP_Non_Dep.ENO;
"""
Constructed_EMP = pd.read_sql(Constructed_EMP_sql, con=mydb)
display(Constructed_EMP)
```

```python
Constructed_EMP_sql="""\
Select EMP_Dep.ENO,ENAME,TITLE,DOB,Dept
From EMP_Dep JOIN EMP_Non_Dep
ON EMP_Dep.ENO = EMP_Non_Dep.ENO;
"""
Constructed_EMP = pd.read_sql(Constructed_EMP_sql, con=mydb)
display(Constructed_EMP)
```

```python
Constructed_EMP_sql="""\
Select ENO,ENAME,TITLE,DOB,Dept
From EMP_Dep NATURAL JOIN EMP_Non_Dep
"""
Constructed_EMP = pd.read_sql(Constructed_EMP_sql, con=mydb)
display(Constructed_EMP)
```

```python

Constructed_EMP_sql="""\
DROP TABLE IF EXISTS con_emp;
CREATE TABLE con_emp as Select ENO,ENAME,TITLE,DOB,Dept
From EMP_Dep NATURAL JOIN EMP_Non_Dep;
"""
mycursor.execute(Constructed_EMP_sql)
```
